```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy.stats import ttest_ind

output_dir ="Output"
os.makedirs(output_dir, exist_ok=True)

files = {
    "Standard SNLI": "PVI/bert-base-cased_snli_std_test.csv",
    "Shuffled SNLI": "PVI/bert-base-cased_snli_test_shuffled.csv",
    "Synonym Noun Word SNLI": "PVI/bert-base-cased_snli_test_modified_noun.csv",
    "Swap Premise Hypothesis SNLI": "PVI/bert-base-cased_snli_test_swap_premise_hypo.csv",
    "Swap Two Halves Premi-Hypo SNLI": "PVI/bert-base-cased_snli_test_swap_two_halves.csv"
}

for name, file_path in files.items():
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"{name} file not found: {file_path}")

datasets = {name: pd.read_csv(file_path) for name, file_path in files.items()}

required_columns = ["PVI", "correct_yx"]
for name, df in datasets.items():
    if not all(col in df.columns for col in required_columns):
        raise KeyError(f"Missing required columns in {name}: {required_columns}")

bins = np.arange(-10, 2, 0.2)

f, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, 1, sharex=True, figsize=(6, 10))

for ax, (name, df) in zip([ax1, ax2, ax3, ax4, ax5], datasets.items()):
    correct = df[df['correct_yx'] == True]['PVI']
    incorrect = df[df['correct_yx'] == False]['PVI']
    t_stat, p_val = ttest_ind(correct, incorrect)
    mean_diff = np.mean(correct) - np.mean(incorrect)
    print(f"{name} - T-test: statistic={t_stat:.3f}, p-value={p_val:.3e}, Mean PVI diff (correct - incorrect)={mean_diff:.3f}")
    ax.hist(correct, weights=np.ones_like(correct)/len(correct), label='correct', alpha=0.5, bins=bins, color='tab:green')
    ax.hist(incorrect, weights=np.ones_like(incorrect)/len(incorrect), label='incorrect', alpha=0.5, bins=bins, color='tab:pink')
    ax.vlines(df['PVI'].mean(), 0, 0.85, linestyles='dashed', color='black')
    ax.set_title(name, fontsize=12)
    ax.set_ylim([0, 0.85])

plt.rcParams['axes.facecolor'] = 'white'
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.7), fontsize=12, ncol=2)
f.subplots_adjust(hspace=0.5)
plt.xlim([-2, 2])
plt.xlabel('PVI (higher $\\leftrightarrow$ easier)', fontsize=14)
ax1.set_ylabel('Proportion of Examples', fontsize=14)
ax1.yaxis.set_label_coords(-0.1, -1.0)
f.patch.set_facecolor('white')
plt.savefig(os.path.join(output_dir, "pvi_snli_variants_comparison.pdf"), dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```
